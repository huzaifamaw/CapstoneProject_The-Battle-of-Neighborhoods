```python
def getNearbyVenues(names, latitudes, longitudes, radius=1000):
    
    venues_list=[]
    i=0
    for name, lat, lng in zip(names, latitudes, longitudes):
        print('Loaded ...' ,i)
        i=i+1
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            1000, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        if(results == []):
            continue
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name,
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Postal Code', 
                  'Longitude',
                   "latitude",         
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
names=df_merged['Postal Code']
lat=df_merged['Latitude']
lng=df_merged['Longitude']
b=df_merged['Borough']
df_ = getNearbyVenues(names[0:20],lat[0:20],lng[0:20])
df_
```

```python
manhattan_onehot = pd.get_dummies(df_[['Venue Category']], prefix="", prefix_sep="")
```

```python
manhattan_onehot['Postal Code']=df_['Postal Code']
```

```python
fixed_columns = [manhattan_onehot.columns[-1]] + list(manhattan_onehot.columns[:-1])
manhattan_onehot = manhattan_onehot[fixed_columns]

manhattan_onehot.head()
```

As we can see we haved added the whole row so to check how many each catagories exists in each postal code!

```python
df_grouped=manhattan_onehot.groupby('Postal Code').sum()
df_grouped.reset_index(inplace=True)
```

```python

kclusters = 6

manhattan_grouped_clustering = df_grouped.drop('Postal Code', 1)


kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(manhattan_grouped_clustering)


kmeans.labels_
```

```python
df_grouped['Clusters']=kmeans.labels_
```
